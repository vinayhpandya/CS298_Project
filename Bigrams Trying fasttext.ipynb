```python
def read_bytes(file_,folder):
    with open('desktop/malware_files_CS297/'+folder+'/'+file_,"rb") as f:
        bytes_=f.read(1)
        while bytes_:
            yield bytes_
            bytes_=f.read(1)
```

```python
from collections import Counter
import nltk
from nltk.util import ngrams
def collect_bytes(folder):
    byte_data=[]
    counter=1
    bigram_data=[]
    for file in os.listdir("desktop/malware_files_CS297/"+folder):
        print(counter)
        str_bytes=""
        for k in read_bytes(file,folder):
            str_bytes+=str(k)+" "
        byte_string=str_bytes.strip()
        counter+=1
        bigram_string=""
        for i in list(byte_string.split())[:20000]:
            bigram_string+=i+" "
        bigram_string=bigram_string.strip()
        bigram_data.append(bigram_string)
        byte_data.append(byte_string)
    return bigram_data
```

```python
Bho_data=collect_bytes("BHO")
print("new")
CeeInject_data=collect_bytes("CeeInject")
print("new")
FakeRean_data=collect_bytes("FakeRean")
print("new")
OnlineGames_data=collect_bytes("OnlineGames")
print("new")
Renos_data=collect_bytes("Renos")
print("new")
Vobfus_data=collect_bytes("Vobfus")
print("new")
Winwebsec_data=collect_bytes("Winwebsec")
```

```python
import numpy as np
```

```python
bigram_array=[]
for k in all_data:
    counter=Counter(k).most_common(20)
    vector_array=[]
    for i,j in counter:
        vector_array.append(model.get_word_vector(i))
    vector_array=np.array(vector_array).reshape(-1)
    bigram_array.append(vector_array)
```

```python
bigram_array=np.array(bigram_array).reshape(len(bigram_array),-1)
```
